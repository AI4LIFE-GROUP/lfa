```python
import numpy as np
import pandas as pd

import sys
sys.path.insert(0, '/n/home07/than157/desktop/meta-algo-v3-regr-wine-norm11') # adding folder with scripts to the system path
from functions_data_processing import center_then_normalize
```

# load clean data

```python
dataset_name = 'who'

test_size = 0.2
train_str = int((1-test_size)*100)
test_str = int(test_size*100)

#load train and test set
filepath = f'datasets/clean/{dataset_name}-clean-train{train_str}.csv'
data_train = pd.read_csv(filepath)

filepath = f'datasets/clean/{dataset_name}-clean-test{test_str}.csv'
data_test = pd.read_csv(filepath)
```

```python
print(f'data_train: {data_train.shape}')
print(f'data_test: {data_test.shape}')
```

# normalize data: subtract mean, then normalize to [0, 1]

```python
#mean-center then normalize data to [0, 1]
#transform X and y
train_normalized, test_normalized = center_then_normalize(data_train, data_test, a=0, b=1)
```

```python
#look at data after mean-centering and normalization
#X_train
display(pd.DataFrame(train_normalized).describe())
display(pd.DataFrame(train_normalized).plot(kind='box'))
#X_test
display(pd.DataFrame(test_normalized).describe())
display(pd.DataFrame(test_normalized).plot(kind='box'))
```

```python
#save data
train_normalized.to_csv(f'datasets/clean/{dataset_name}-clean-train{train_str}-normalized.csv', index=False)
test_normalized.to_csv(f'datasets/clean/{dataset_name}-clean-test{test_str}-normalized.csv', index=False)
```
