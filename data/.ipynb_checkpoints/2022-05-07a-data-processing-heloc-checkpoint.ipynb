```python
import pandas as pd
import numpy as np
```

# load data

```python
dataset_name = 'heloc'
```

```python
#from chirag's benchmark datasets
train = pd.read_csv('original/heloc-train.csv', sep=',')
test = pd.read_csv('original/heloc-test.csv', sep=',')
```

```python
train.shape
```

```python
test.shape
```

# clean data

```python
train.isna().sum()
```

```python
test.isna().sum()
```

### combine train and test set and resplit (test set has an outlier)

```python
data_all = pd.concat([train, test], axis=0)
```

```python
data_all
```

```python
#save full dataset
data_all.to_csv(f'clean/{dataset_name}-clean-full.csv', index=False)
```

# split data into train/test

```python
from sklearn.model_selection import train_test_split
```

```python
#split data
test_size = 0.20
data_train, data_test = train_test_split(data_all, test_size=test_size, random_state=42)

#save data
train_str = int((1-test_size)*100)
test_str = int(test_size*100)
data_train.to_csv(f'clean/{dataset_name}-clean-train{train_str}.csv', index=False)
data_test.to_csv(f'clean/{dataset_name}-clean-test{test_str}.csv', index=False)
```

# normalize data: subtract mean, then normalize to [0, 1]

```python
def center_then_normalize(X_train, X_test, a, b):
    #mean center X matrix
    train_means = X_train.mean()
    X_train_new = X_train - train_means
    X_test_new = X_test - train_means

    #normalize mean-centered data to [-1, 1]
    train_mins = X_train_new.min(axis=0)
    train_maxs = X_train_new.max(axis=0)

    #normalize X_train and X_test to range [a, b] = [-1, 1]
    X_train_new = (b-a)*((X_train_new-train_mins)/(train_maxs-train_mins)) + a
    X_test_new = (b-a)*((X_test_new-train_mins)/(train_maxs-train_mins)) + a
    
    return X_train_new, X_test_new
```

```python
#mean-center then normalize data to [0, 1]
#transform X and y
train_normalized, test_normalized = center_then_normalize(data_train, data_test, a=0, b=1)
```

```python
#look at data after mean-centering and normalization
#X_train
display(pd.DataFrame(train_normalized).describe())
display(pd.DataFrame(train_normalized).plot(kind='box'))
#X_test
display(pd.DataFrame(test_normalized).describe())
display(pd.DataFrame(test_normalized).plot(kind='box'))
```

```python
#save data
train_normalized.to_csv(f'clean/{dataset_name}-clean-train{train_str}-normalized.csv', index=False)
test_normalized.to_csv(f'clean/{dataset_name}-clean-test{test_str}-normalized.csv', index=False)
```
